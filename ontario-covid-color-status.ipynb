```python
import urllib
import pandas as pd
pd.set_option('display.max_rows', 500)
import requests
import json
import numpy as np
```

```python
# Datasource URL: https://data.ontario.ca/dataset/ontario-covid-19-zones/resource/ce9f043d-f0d4-40f0-9b96-4c8a83ded3f6
url='https://data.ontario.ca/api/3/action/datastore_search?resource_id=ce9f043d-f0d4-40f0-9b96-4c8a83ded3f6'
url_limit=url+'&limit='
data=requests.get(url)
json_object=json.loads(data.content)
limit=json_object['result']['total']

data_total=requests.get(url_limit+str(limit))
json_object_total=json.loads(data_total.content)

#column_names
```

```python
data_points=[]
for row in json_object_total['result']['records']:
    data_points.append(row)
```

```python
color_status = pd.DataFrame(data_points)
reporting_phu = color_status['Reporting_PHU'].drop_duplicates().tolist()

#print(reporting_phu)

color_status['Active'] = 'Not Active'

color_status

for phu in reporting_phu:
    temp = color_status[color_status['Reporting_PHU'] == phu]
    max_id = temp['_id'].max()
    max_id = max_id - 1
    color_status.loc[max_id, 'Active'] = 'Active'
```

```python
color_status.to_csv('ontario-covid-color-status.csv')
```
