```python
import urllib
import pandas as pd
pd.set_option('display.max_rows', 500)
import requests
import json
```

```python
# Datasource URL: https://data.ontario.ca/api/3/action/datastore_search?resource_id=8a89caa9-511c-4568-af89-7f2174b4378c
url='https://data.ontario.ca/api/3/action/datastore_search?resource_id=8a89caa9-511c-4568-af89-7f2174b4378c'
url_limit=url+'&limit='
data=requests.get(url)
json_object=json.loads(data.content)
limit=json_object['result']['total']

data_total=requests.get(url_limit+str(limit))
json_object_total=json.loads(data_total.content)
```

```python
data_points=[]
for row in json_object_total['result']['records']:
    data_points.append(row)
```

```python
data_points
```

```python
pd.DataFrame(data_points).to_csv('ontario-vaccine-progress.csv')
```
